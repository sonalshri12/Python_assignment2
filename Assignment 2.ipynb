```python
# Task 1: Find the largest of three numbers
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

if num1 >= num2 and num1 >= num3:
    print(f"{num1} is the largest.")
elif num2 >= num1 and num2 >= num3:
    print(f"{num2} is the largest.")
else:
    print(f"{num3} is the largest.")
```

```python
# Task 2: Check if a number is even or odd
num = int(input("Enter a number: "))
if num % 2 == 0:
    print(f"{num} is even.")
else:
    print(f"{num} is odd.")
```

```python
# Task 3: Print numbers from 1 to 100, skipping divisible by 7
for i in range(1, 101):
    if i % 7 == 0:
        continue
    print(i, end=" ")
```

```python
# Task 4: Print multiplication table of a given number
num = int(input("Enter a number: "))
for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
```

```python
# Task 5: Print a right-angled triangle pattern of stars
rows = int(input("Enter the number of rows: "))
for i in range(1, rows + 1):
    print("*" * i)
```

```python
# Task 6: Print the pattern
rows = int(input("Enter the number of rows: "))
for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print(j, end="")
    print()
```

```python
# Task 7: Check if a character is a vowel or consonant
char = input("Enter a character: ").lower()
if char in ['a', 'e', 'i', 'o', 'u']:
    print(f"{char} is a vowel.")
else:
    print(f"{char} is a consonant.")
```

```python
# Task 8: Print a diamond pattern of stars
n = int(input("Enter the number of rows (odd number): "))
for i in range(1, n + 1, 2):
    print(" " * ((n - i) // 2) + "*" * i)
for i in range(n - 2, 0, -2):
    print(" " * ((n - i) // 2) + "*" * i)
```

```python
# Task 9: Reverse the digits of a number
num = int(input("Enter a number: "))
reversed_num = int(str(num)[::-1])
print(f"Reversed number: {reversed_num}")
```

```python
# Task 10: Print the pyramid pattern
rows = int(input("Enter the number of rows: "))
for i in range(1, rows + 1):
    # Print spaces
    print(" " * (rows - i), end="")
    # Print increasing numbers
    for j in range(1, i + 1):
        print(j, end="")
    # Print decreasing numbers
    for j in range(i - 1, 0, -1):
        print(j, end="")
    print()
```
